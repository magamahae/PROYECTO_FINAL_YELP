## Web Scraping 

Importamos librerias necesarias

```python
import requests
import numpy as np
import herramientas
from IPython.display import Image, display
from bs4 import BeautifulSoup
import pandas as pd
```

Por medio del siguiente codigo, extraigo la tabla de "Wikipedia" y la guardo en un DataFrame 

```python

url= 'https://es.wikipedia.org/wiki/Anexo:Estados_de_los_Estados_Unidos_por_PIB'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    table = soup.find('table', {'class': 'wikitable'})
    
    if table:
        rows = table.find_all('tr')
        
        data = []  # Lista para almacenar filas de datos
        
        for row in rows:
            cells = row.find_all(['th', 'td'])
            row_data = []  # Lista para almacenar datos de cada fila
            
            for cell in cells:
                row_data.append(cell.text.strip())
            
            if row_data:  # Asegurarse de que la fila tenga datos antes de agregarla
                data.append(row_data)
        
        # Crear un DataFrame con los datos recolectados
        df = pd.DataFrame(data[1:], columns=data[0])
        
        # Mostrar el DataFrame
        print(df)
        
        # También puedes guardar el DataFrame en un archivo CSV
        df.to_csv('tabla_extraida.csv', index=False)  # Esto guarda el DataFrame en un archivo CSV
        
    else:
        print('No se encontró la tabla con la clase especificada.')
else:
    print('Error al obtener la página:', response.status_code)
```
