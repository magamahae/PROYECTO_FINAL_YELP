Importamos las librerias necesarias 

```python
import os
import json 
import numpy as np
import pandas as pd
import herramientas
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
```

Leemos los archivos .csv que obtuvimos del ETL 

```python
df_california=pd.read_csv(r'california.csv')
df_florida=pd.read_csv(r'florida.csv')
df_illinois=pd.read_csv(r'illinois.csv')
df_texas=pd.read_csv(r'texas.csv')
df_new_york=pd.read_csv(r'new_york.csv')
df_google_review=pd.read_csv(r'df_sitios_estados.csv')
```

- **Google: Sitios Estados** 

Creamos un filtro para que solo trabajemos con los que tengan la palabra 'restaurant' en la columna category

```python
filtro_comida = df_google_review['category'].str.contains('restaurant', case=False)
#Crear un nuevo DataFrame con las filas relacionadas con comida
df_comida_g = df_google_review[filtro_comida]

#Imprimir el nuevo DataFrame
df_comida_g
```

```python
df_comida_g.columns
```

Eliminamos las columnas que no utilizaremos ('address','MISC','relative_results','url') 

```python
columns=['address','MISC','relative_results','url']
df_comida_g_Clean= df_comida_g.drop(columns, axis=1)
df_comida_g_Clean
```

Eliminamos los corchetes y las comillas de la columna 'category', creando la siguiente funcion, la cual crea una nueva columna 'categoria_extraida'

```python
#Define una función para procesar cada valor de la columna
def procesar_categoria(categoria):
    if pd.notna(categoria):
        # Elimina los corchetes y comillas y separa las categorías por comas
        categorias_lista = [cat.strip("'") for cat in categoria[1:-1].split(',')]
        return ','.join(categorias_lista)
    else:
        return None

#Aplica la función a la columna 'categoria' y crea una nueva columna 'categoria_extraida'
df_comida_g_Clean['categoria_extraida'] = df_comida_g_Clean['category'].apply(procesar_categoria)

#Imprime el DataFrame con la nueva columna
df_comida_g_Clean
```

Eliminamos comillas de la nueva columna creando una nueva llamada 'categoria_extraida_1'

```python
df_comida_g_Clean['categoria_extraida_1']=df_comida_g_Clean['categoria_extraida'].replace("'","",regex=True)
df_comida_g_Clean
```

Eliminamos las columnas creadas anteriormente, y renombraremos la columna 

```python
columns=['category','categoria_extraida']
df_comida_g_Clean=df_comida_g_Clean.drop(columns,axis=1)
df_comida_g_Clean
```

```python
df_comida_g_Clean.rename(columns={'categoria_extraida_1': 'category'}, inplace=True)
df_comida_g_Clean
```

Exportamos a un archivo .csv 

```python
# Exportar el DataFrame
df_comida_g_Clean.to_csv('business_google.csv', escapechar='\\', index=False)
```

Concatenamos los dataframes de los estados en un solo dataframe

```python
#Combina los DataFrames usando la función concat
df_reviews = pd.concat([df_florida, df_texas, df_illinois, df_new_york, df_california], ignore_index=True)

#Imprime el nuevo DataFrame
df_reviews
```

```python
#Filtra las filas de df_reviews con gmap_id que coincidan con los de df_comida_g
df_reviews_filtrado = df_reviews[df_reviews['gmap_id'].isin(df_comida_g['gmap_id'])]

#Imprime el nuevo DataFrame con las filas filtrad
df_reviews_filtrado
```

```python
df_reviews_filtrado.columns
```

```python
df_reviews_filtrados=df_reviews_filtrado.drop('name', axis=1)
df_reviews_filtrados
```

```python
# Exportar el DataFrame
df_reviews_filtrado.to_csv('reviews_google.csv', escapechar='\\', index=False)
```
