# Esto hace análisis de sentimiento y se lo incorpora a las reviews de YELP

```python
import pandas as pd
```

```python
reviews_yelp_restaurant = pd.read_json(r'ETL/Archivos para unir/reviews_yelp_restaurant_reducido.json')
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Defines a categorization for each sentiment analyzed: 2 for positive, 1 for neutral and 0 for negative.
def categorize_sentiment(score):
    if score < -0.05:
        return 0 #Bad
    elif score > 0.05:
        return 2 #Positive
    else:
        return 1 #Neutral
# In case it's needed, we make sure that the column is a string
reviews_yelp_restaurant['text'] = reviews_yelp_restaurant['text'].astype(str) 
```

```python
# Applys the sentintiment analysis to reviews and sorts results
reviews_yelp_restaurant['pol_score'] = reviews_yelp_restaurant['text'].apply(lambda review: sia.polarity_scores(review)['compound'])
```

```python
reviews_yelp_restaurant
```

```python
reviews_yelp_restaurant['rating'] = reviews_yelp_restaurant['pol_score'].apply(categorize_sentiment)
```
