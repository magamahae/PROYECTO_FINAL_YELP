```python
#Análisis de sentimiento para yelp y google actualizado, luego de algunos cambios en los datos por parte del departamento de data analytics.
#Importar librerias:
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from google.colab import drive
# ---------------------------------------------------------------
#Montar Drive y descargar Vader Lexicon:
drive.mount('/content/drive')
nltk.download('vader_lexicon')
# ---------------------------------------------------------------
#Crear dataframe:
df_yelp = pd.read_json('/content/drive/My Drive/PRUEBA COLAB/ETL/Archivos para unir/reviews_yelp_restaurant_reducido.json')
#Guardar el analisis aplicado a la columna 'text' en una variable:
sia = SentimentIntensityAnalyzer()
# ---------------------------------------------------------------
#Crear una nueva columna con el analisis de sentimiento:
df_yelp['S_A'] = df_yelp['text'].apply(lambda x: sia.polarity_scores(str(x))['compound'])
# ---------------------------------------------------------------
#Crear categorías para los rangos de la columna 'S_A':
df_yelp['sentiment_category'] = df_yelp['S_A'].apply(lambda x: 'Positivo' if x > 0 else ('Negativo' if x < 0 else 'Neutral'))
# ---------------------------------------------------------------
#Nuevo dataframe con la columna 'S_A' y 'sentiment_category':
print(df_yelp[['text', 'S_A', 'sentiment_category']])
```

```python
#Guardar el analisis aplicado a la columna 'text' en una variable:
sia2 = SentimentIntensityAnalyzer()
# ---------------------------------------------------------------
#Crear una nueva columna con el analisis de sentimiento:
df_google['S_A'] = df_google['text'].apply(lambda x: sia2.polarity_scores(str(x))['compound'])
# ---------------------------------------------------------------
#Crear categorías para los rangos de la columna 'S_A':
df_google['sentiment_category'] = df_google['S_A'].apply(lambda x: 'Positivo' if x > 0 else ('Negativo' if x < 0 else 'Neutral'))
# ---------------------------------------------------------------
#Nuevo dataframe con la columna 'S_A' y 'sentiment_category':
print(df_google[['text', 'S_A', 'sentiment_category']])
```
